```python
import numpy as np
def simulate():
    num_samples = 10000
    lambda_m = np.array([[0, 1], [3, 0]])  # define lambda matrix
    prior_prob = np.array([1 / 4, 3 / 4])  # define prior probability
    m1 = 2
    s1 = np.sqrt(0.5)
    m2 = 1.8
    s2 = np.sqrt(0.2)
    # generate random samples
    sample_1 = np.random.normal(m1, s1, num_samples)
    sample_2 = np.random.normal(m2, s2, num_samples)
    overal_cost=0

    # initialize cost vars
    for i in range(num_samples):
        posterior_1 = np.exp(-((sample_1[i] - m1) ** 2) / (2 * s1 ** 2)) / (np.sqrt(m1 * np.pi) * s1)
        posterior_2 = np.exp(-((sample_2[i] - m2) ** 2) / (2 * s2 ** 2)) / (np.sqrt(m2 * np.pi) * s2)

        p=posterior_1*prior_prob[0] + posterior_2 * prior_prob[1]
        posterior_1=(posterior_1*prior_prob[0])/p
        posterior_2=(posterior_2*prior_prob[1])/p


        cost_1 = lambda_m[0, 0] * posterior_1 + lambda_m[0, 1] * posterior_2
        cost_2 = lambda_m[1, 0] * posterior_1 + lambda_m[1, 1] * posterior_2
        overal_cost+=(cost_1+cost_2)
        # if cost_1<cost_2:
        #     risk=cost_1
        # else :
        #     risk=cost_2

    return(overal_cost,overal_cost/num_samples)

cost=simulate()
print(f"The overal cost is: {cost[0]}\n")
print(f"The mean_overal cost is :{cost[1]}")
```

      - ${g_i(x)=-\frac{1}{2}(x-μ_i)^T \ Σ_i^{-1}(x-μ_i)-\frac{d}{2}ln(2π)-\frac{1}{2}ln(|Σ_i|)+ ln(P(ω_i)) }$ <br />

       Η κατανομή υλοποιείται από την παρακάτων συνάρτηση "disc_fnc".<br/>
       
       Σημείωση: Για $d=1$ η συνάρτηση διάκρισης μετατρέπεται σε :<br/>
       - $ g_i(x)=-\frac{1}{2}\frac{(x-μ_i)^2}{σ_i^2}-\frac{1}{2}ln(2πσ_i^2)+ln(P(ω_i))$
   * ii ) Η ευκλείδεια απόσταση δύο σημείων ${x_1,x_2}$ σε ένα χώρο d διαστάσεων ορίζεται ως:<br />
        - ${d(x_1,x_2)=\sqrt{ \sum_{i=1}^{d}(x_{1i}-x_{2i})^2}}$<br />
  
      και υλοποιείται από την "euclid_distance"
   * iii )  Η απόσταση Mahalanobis μεταξύ του μέσου $μ$ και ενός αυθαίρετου σημείου $x$ σε χώρο $d$ διαστάσεων ,δεδομένου του πίνακα συνδιασποράς ορίζεται ως :<br />
        - ${D_M(x,μ)=\sqrt{ (x-μ)^{Τ} \ Σ^{-1}(x-μ)}}$<br />
        και υλοποιείται από την "mahalanobis_distance".

   


```python
import numpy as np

def discriminant_function(x, class_mean, class_covariance, class_prior):


    d = x.shape[0]  # Dimension of the feature vectors

    # Calculate the inverse of the class covariance matrix
    class_covariance_inv = np.linalg.inv(class_covariance)

    # Term 1: Quadratic term
    term_1 = -0.5 * np.dot(np.dot((x - class_mean).T, class_covariance_inv), (x - class_mean))

    # Term 2: Constant term related to dimension
    term_2 = -0.5 * d * np.log(2 * np.pi)

    # Term 3: Logarithm of determinant of covariance matrix
    term_3 = -0.5 * np.log(np.linalg.det(class_covariance))

    # Term 4: Logarithm of the class prior probability
    if class_prior==0:
        term_4=0
    else:
       term_4 = np.log(class_prior)

    # Calculate the discriminant function
    discriminant_value = term_1 + term_2 + term_3 + term_4

    return discriminant_value

def euclid_distance(x1,x2):
    squared = [(a - b) ** 2 for a, b in zip(x1, x2)]
    distance = np.sqrt(sum(squared))
    return distance

def mahalanobis_distance(x,mean,covariance):

    inv_covariance=np.linalg.inv(covariance)

    diff=x-mean

    distance=np.sqrt(np.dot(np.dot(diff.T,inv_covariance),diff))

    return(distance)
```

```python
import numpy as np
def calculate_variance_mean(values):
    # calculate mean value
    mean_list=[]
    var_list=[]
    filtered_1 = data[data['ω_i'] == 1]
    mean_1=filtered_1[values].mean()
    var_1=filtered_1[values].var()
    mean_list.append(mean_1)
    var_list.append(var_1)

    filtered_2 = data[data['ω_i'] == 2]
    mean_2 = filtered_2[values].mean()
    var_2 = filtered_2[values].var()
    mean_list.append(mean_2)
    var_list.append(var_2)

    filtered_3 = data[data['ω_i'] == 3]
    mean_3 = filtered_3[values].mean()
    var_3 = filtered_3[values].var()
    mean_list.append(mean_3)
    var_list.append(var_3)


    return(mean_list,var_list)

def covariance(x1,x2,mean_x1,mean_x2,number):

    filtered_x1=data[data["ω_i"]==number][x1]
    filtered_x2=data[data["ω_i"]==number][x2]
    # devi_x1=[xi-mean_x1 for xi in filtered_x1]
    # devi_x2=[xi-mean_x2 for xi in filtered_x2]
    # product=[dx1*dx2 for dx1,dx2 in zip(devi_x1,devi_x2)]
    # final_cov=sum(product)/len(data)
    final_cov=np.cov(filtered_x1,filtered_x2)[0,1]
    return(final_cov)


def cov_matrix(mean_matrix,var_matrix):
    cov_matrix_1=[]
    cov_matrix_2=[]
    cov_matrix_3=[]
    x1="x1"
    x2="x2"
    x3="x3"
    a=0
    b=1

    while (a<3):
        cov_1 = covariance(x1, x2, mean_matrix[0][a], mean_matrix[1][a], b)

        if mean_matrix.shape[0]==3:
           cov_2=covariance(x2,x3,mean_matrix[1][a],mean_matrix[2][a],b)
           cov_3=covariance(x1,x3,mean_matrix[0][a],mean_matrix[2][a],b)
           temp=[
               [var_matrix[0][a],cov_1,cov_3],
               [cov_1,var_matrix[1][a],cov_2],
               [cov_3,cov_2,var_matrix[2][a]]
           ]
        else:
            temp = [
                [var_matrix[0][a], cov_1],
                [cov_1, var_matrix[1][a]]
            ]

        if(b==1):
            cov_matrix_1=temp
        if(b==2):
            cov_matrix_2=temp
        if(b==3):
            cov_matrix_3=temp
        else:
            cov_matrix_1=cov_matrix_1
            cov_matrix_2=cov_matrix_2
            cov_matrix_3=cov_matrix_3
        a = a+1
        b = b+1

        cov_matrix_1=np.array(cov_matrix_1)
        cov_matrix_2=np.array(cov_matrix_2)
        cov_matrix_3=np.array(cov_matrix_3)




    return(cov_matrix_1,cov_matrix_2,cov_matrix_3)
```
